```python
import face_recognition
import json
import pickle
import numpy as np
import os

# Load the CNN face detection model
face_detection_method = "hog"

# Initialize lists to store face encodings and corresponding names
known_encodings = []
known_names = []

# Loop through each person's directory and load their face images
for person_dir in os.listdir('N18'):
    if not os.path.isdir(os.path.join('N18', person_dir)):
        continue

    person_images = []
    for image_file in os.listdir(os.path.join('N18', person_dir)):
        image = face_recognition.load_image_file(os.path.join('N18', person_dir, image_file))
        face_locations = face_recognition.face_locations(image, model=face_detection_method)
        if len(face_locations) == 0:
            print("No faces found in image: {}".format(os.path.join('N18', person_dir, image_file)))
            continue
        encodings = face_recognition.face_encodings(image, known_face_locations=face_locations)
        for encoding in encodings:
            # Reshape each encoding to have the same shape before flattening
            flattened_encoding = np.array(encoding).reshape(-1)
            person_images.append(flattened_encoding)

    known_encodings.append(np.array(person_images))
    known_names.append(person_dir)

# Print the number of known faces
print("Number of known faces: {}".format(len(known_names)))

# Export the trained model
model_data = {
    "known_encodings": known_encodings,
    "known_names": known_names,
    "face_detection_method": face_detection_method
}
print(model_data)
with open("face_recognition_model.pkl", "wb") as f:
    pickle.dump(model_data, f)
```
